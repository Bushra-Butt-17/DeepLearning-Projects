# <center><h1> Data Loading and Inspection</h1></center>


```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

 <h3>1. Loading the Iris dataset into a pandas DataFrame</h3><br>
 Loading the dataset using the seaborn library by the 'load_dataset' function and 'iris' as its argument.

```python
df = sns.load_dataset('iris')
```

<h3>2. Displaying the first 5 and the last 5 rows of the dataset.</h3><br>
To display the first 5 rows df.head is used while to display the last 5 rows df.tail is used.

```python
print("First 5 rows of the Iris dataset:")
print(df.head(), "\n") 
print("Last 5 rows of the Iris dataset:")
print(df.tail())
```

<h3>3. Printing the summary statistics and checking for any missing/null values in the dataset. </h3><br>
To get a summary of numerical data, df.describe() is used. This method provides a quick overview of central tendency, dispersion, and distribution shape, including metrics like count, mean, standard deviation, minimum, maximum, and various percentiles.<br>
To identify if any values are missing or null,df.isnull() can be used, it checks each column for missing or null values and returns the total count of nulls per column. Since all results are zero, it means there are no missing values in the dataset.

```python
# Summary statistics
print("Summary statistics for the Iris dataset:")
print(df.describe())

# Check for missing or null values
print("\nChecking for missing or null values in the Iris dataset:")
print(df.isnull().sum())
```

<h3> 4. Displaying the datatype of each column.</h3><br>
To display the data type of each column in a Pandas DataFrame, dtypes attribute is used. It provides information about the data type for each column, which can be useful when analyzing or preprocessing data.

```python
print("Data types for each column in the Iris dataset:")
print(df.dtypes)
```

# <center><h1> Data Manipulation</h1></center>


<h3>1. Adding a new column named 'sepal_area' that is the product of sepal length and sepal width.</h3><br>
Multiplying these two existing columns and assigning the result to a new column in the DataFrame.

```python
df['sepal_area']=df['sepal_length']*df['sepal_width']
print("First 5 rows with 'sepal_area':")
print(df.head())
```

<h3>2. Creating a new DataFrame that contains only the data for the 'setosa' species.</h3><br>
The expression below creates a boolean mask that identifies which rows have species equal to 'setosa'. The resulting boolean mask is used to filter the DataFrame, creating setosa_df with only the rows where species is 'setosa'. Then, to confirm the new DataFrame contains only 'setosa' data, the first few rows using head() are displayed.

```python
setosa_df=df[df['species']=='setosa']
print("Data for 'setosa' species:")
print(setosa_df.head())
```

<h3>3. Sorting the original DataFrame by 'petal_length' in descending order.</h3><br>
The sort_values(by='petal_length', ascending=False) line sorts the DataFrame by 'petal_length' in descending order. To keep the original DataFrame intact, the result of sort_values() is assigned to a new variable, sorted_iris_df.Then, to verify the sorting, the first few rows of sorted_df are displayed using head().

```python
sorted_df = df.sort_values(by='petal_length',ascending=False)
print("Iris dataset sorted by 'petal_length' in descending order:")
print(sorted_df.head())
```

<h3>4. Filtering the data to include only those records with a petal width greater than 1.5 cm.</h3><br>
The condition df['petal_width'] > 1.5 creates a boolean mask that identifies rows where 'petal_width' is greater than 1.5 cm. The boolean mask is used to filter the DataFrame, resulting in a new DataFrame filtered_df containing only the records that meet the condition. The resulting dataset is displayed to confirm the filtering worked or not.

```python
filtered_df=df[df['petal_width']>1.5]
print("Records with petal width greater than 1.5 cm:")
print(filtered_df.head())
```

<h3>5. Calculate the mean sepal width for each of the species present in the dataset. </h3><br>
The df.groupby('species') creates groups based on unique values in the 'species' column. After grouping, the mean for 'sepal_width' is calculated using .mean(). This calculates the average sepal width for each group of species. The result is a Pandas Series with 'species' as the index and the calculated mean for 'sepal_width'.

```python
mean_sepal_width = df.groupby('species')['sepal_width'].mean()
print("Mean sepal width for each species:")
print(mean_sepal_width)
```

# <center><h1> Data Visualization</h1></center>


<h4>1. Plotting a histogram of the 'sepal_length' for the entire dataset and Customizing the plot by adding a title and 
labels for x and y axes.</h4>
<ul>
  <li><strong>plt.figure(figsize=(8, 6))</strong>: Adjusts the plot size. The <strong>figsize</strong> parameter controls the width and height of the plot in inches.</li>
  <li><strong>sns.histplot(...)</strong>: Seaborn's function for plotting histograms.</li>
  <ul>
    <li><strong>data=iris_df['sepal_length']</strong>: The data to plot in the histogram, specifically the 'sepal_length' column from the Iris dataset.</li>
    <li><strong>kde=True</strong>: Adds a Kernel Density Estimate (a smoothed curve) to the histogram, showing the data distribution more smoothly.</li>
    <li><strong>bins=20</strong>: Specifies the number of bins (bar sections) for the histogram, influencing the level of granularity.</li>
    <li><strong>color='skyblue'</strong>: Sets the color of the histogram bars. You can use any valid color or hex code.</li>
  </ul>
  <li><strong>plt.title("Histogram of Sepal Length")</strong>: Adds a title to the histogram plot.</li>
  <li><strong>plt.xlabel("Sepal Length (cm)")</strong> and <strong>plt.ylabel("Frequency")</strong>: Label the x-axis and y-axis, respectively.</li>
  <li><strong>plt.show()</strong>: Renders and displays the plot.</li>
</ul>


```python
# Plotting a histogram of 'sepal_length'
plt.figure(figsize=(8, 6))  # Optional: adjust the plot size
sns.histplot(df['sepal_length'], kde=True, bins=20, color='skyblue')  # Histogram with KDE and custom bins
plt.title("Histogram of Sepal Length")  # Add a title
plt.xlabel("Sepal Length (cm)")  # Label for the x-axis
plt.ylabel("Frequency")  # Label for the y-axis

# Show the plot
plt.show()
```

<h4>2. Creating a scatter plot to visualize the relationship between 'sepal_length' and 'sepal_width' and Coloring the 
points by the species of the iris.</h4>
<ul>
  <li><strong>plt.figure(figsize=(8, 6))</strong>: Adjusts the plot size. The <strong>figsize</strong> parameter controls the plot's width and height in inches.</li>
  <li><strong>sns.scatterplot(...)</strong>: Seaborn's scatter plot function to create scatter plots.</li>
  <ul>
    <li><strong>data=iris_df</strong>: Specifies the DataFrame containing the data for the plot.</li>
    <li><strong>x='sepal_length'</strong> and <strong>y='sepal_width'</strong>: Determine which columns to use for the x-axis and y-axis, respectively.</li>
    <li><strong>hue='species'</strong>: Colors the scatter plot points based on the categorical variable 'species'.</li>
    <li><strong>palette='viridis'</strong>: Specifies the color palette for the plot.</li>
    <li><strong>s=50</strong>: Sets the size of the scatter plot markers (higher values mean larger markers).</li>
  </ul>
  <li><strong>plt.title("Scatter Plot of Sepal Length vs. Sepal Width")</strong>: Adds a title to the scatter plot.</li>
  <li><strong>plt.xlabel("Sepal Length (cm)")</strong> and <strong>plt.ylabel("Sepal Width (cm)")</strong>: Set the labels for the x and y axes, respectively.</li>
  <li><strong>plt.legend(title='Species')</strong>: Customizes the legend to indicate what the colors represent.</li>
  <li><strong>plt.show()</strong>: Renders and displays the plot.</li>
</ul>



```python
# Create a scatter plot with 'sepal_length' and 'sepal_width', colored by 'species'
plt.figure(figsize=(8, 6))  # Optional: adjust the plot size
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='viridis', s=50)
plt.title("Scatter Plot of Sepal Length vs. Sepal Width")  # Add a title
plt.xlabel("Sepal Length (cm)")  # Label for the x-axis
plt.ylabel("Sepal Width (cm)")  # Label for the y-axis
plt.legend(title='Species')  # Customize the legend

# Show the plot
plt.show()
```

<h4>3. Using a boxplot to show the distribution of 'petal_length' for each species and Customizing this plot by adding a 
title and axis labels. </h4>
<ul>
  <li><strong>plt.figure(figsize=(8, 6))</strong>: Adjusts the plot size. The <strong>figsize</strong> parameter controls the width and height of the plot in inches.</li>
  <li><strong>sns.boxplot(...)</strong>: Seaborn's function for creating boxplots.</li>
  <ul>
    <li><strong>data=iris_df</strong>: Specifies the data source for the plot (the Iris dataset).</li>
    <li><strong>x='species'</strong>: Defines the categorical variable for the x-axis (grouping by species).</li>
    <li><strong>y='petal_length'</strong>: Specifies the numerical variable for the y-axis (the values to be plotted).</li>
    <li><strong>palette='pastel'</strong>: Sets the color palette for the boxplots, providing a softer color scheme.</li>
  </ul>
  <li><strong>plt.title("Distribution of Petal Length by Species")</strong>: Adds a title to the boxplot, describing the plot's purpose.</li>
  <li><strong>plt.xlabel("Species")</strong> and <strong>plt.ylabel("Petal Length (cm)")</strong>: Set the labels for the x and y axes, respectively.</li>
  <li><strong>plt.show()</strong>: Renders and displays the plot to visualize the boxplots.</li>
</ul>


```python
# Create a boxplot for 'petal_length' grouped by 'species'
plt.figure(figsize=(8, 6))  # Adjusts the plot size

sns.boxplot(  # Seaborn's boxplot function
    data=df,  # The data source
    x='species',  # The categorical variable for the x-axis (grouping by 'species')
    y='petal_length',  # The numerical variable for the y-axis
    palette='pastel'  # Sets the color palette for the boxplots
)

# Customization of the plot
plt.title("Distribution of Petal Length by Species")  # Adds a title to the plot
plt.xlabel("Species")  # Sets the label for the x-axis
plt.ylabel("Petal Length (cm)")  # Sets the label for the y-axis

# Render the plot
plt.show()
```

<h4>4. Create a pairplot for all numerical columns, colored by species. </h4>

<ul>
  <li><strong>sns.pairplot(...)</strong>: Seaborn's function for creating pairplots, which generates scatter plots for each pairwise combination of numerical columns.</li>
  <ul>
    <li><strong>data=iris_df</strong>: Specifies the DataFrame for the pairplot, which in this case is the Iris dataset.</li>
    <li><strong>hue='species'</strong>: Colors the scatter plot points by the categorical variable 'species'.</li>
    <li><strong>palette='viridis'</strong>: Specifies the color palette for the pairplot. Viridis is a popular palette that is colorblind-friendly.</li>
    <li><strong>diag_kind='kde'</strong>: Uses Kernel Density Estimation (KDE) for the diagonal plots to visualize distribution.</li>
    <li><strong>markers=['o', 's', 'D']</strong>: Sets different marker shapes for each species, providing visual differentiation. 'o' is a circle, 's' is a square, and 'D' is a diamond.</li>
  </ul>
  <li><strong>plt.suptitle("Pairplot of Iris Dataset by Species", y=1.02)</strong>: Adds a title to the pairplot. The <strong>y=1.02</strong> adjusts the title's position to prevent overlap with the plots.</li>
  <li><strong>plt.show()</strong>: Renders and displays the pairplot, allowing you to visualize the scatter plots and diagonal distributions.</li>
</ul>


```python
# Create a pairplot for all numerical columns, colored by 'species'
sns.pairplot(  # Seaborn's pairplot function
    data=df,  # The DataFrame containing the data
    hue='species',  # Color the points by 'species'
    palette='viridis',  # Set the color palette
    diag_kind='kde',  # Use Kernel Density Estimation on the diagonal plots
    markers=['o', 's', 'D'],  # Set marker types for different species
)

# Add a title to the pairplot
plt.suptitle("Pairplot of Iris Dataset by Species", y=1.02)  # Adjust the title position

# Display the pairplot
plt.show()
```

<h4>Explanation of Histogram for Sepal Length with Species Hue</h4>
<ul>
 <li><strong>Creating the Histogram</strong>:
    <ul>
      <li><strong>plt.figure(figsize=(8, 6))</strong>: Adjusts the plot size to 8 inches wide and 6 inches tall.</li>
      <li><strong>sns.histplot(data=iris_df, x='sepal_length', hue='species', multiple='stack', bins=20, palette='viridis', kde=True)</strong>:
        <ul>

          <li><strong>kde=True</strong>: Adds a Kernel Density Estimate, a smoothed curve indicating the data's distribution.</li>
        </ul>

    </ul>

    </ul>
  </li>

  <li><strong>Rendering the Histogram</strong>:
    <ul>
      <li><strong>plt.show()</strong>: Renders and displays the histogram.</li>
    </ul>
  </li>

```python
plt.figure(figsize=(8, 6))  # Adjust the plot size
sns.histplot(data=df, x='sepal_length', hue='species', multiple='stack', kde=True, bins=20, palette='viridis')  # Stacked histogram with KDE
plt.title("Histogram of Petal Length by Species")  # Add a title
plt.xlabel("Sepal Length (cm)")  # Label for the x-axis
plt.ylabel("Count")  # Label for the y-axis
plt.show()  # Display the plot
```
